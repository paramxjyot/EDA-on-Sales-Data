```python
#top highest performing cities wrt to states

df_sorted = df.sort_values('Lineitem quantity', ascending=False)
df_grouped = df_sorted.groupby(['Ship State', 'Shipping City'])['Lineitem quantity'].size()
result = df_grouped.groupby('Ship State').nlargest(3).droplevel(0)

print(result)

```

```python
#top 3 lowest performing cities wrt to states
df_sorted = df.sort_values('Lineitem quantity', ascending=False)
df_grouped = df_sorted.groupby(['Ship State', 'Shipping City'])['Lineitem quantity'].size()
result_lowest = df_grouped.groupby('Ship State').nsmallest(3).droplevel(0)
print(result_lowest)
```
